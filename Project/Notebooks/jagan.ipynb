```python
import pandas as pd

import wine_main

wine_main.wineDF.head()

import matplotlib.pyplot as plt

import numpy as np
```

```python
#Create type of wine DF
wine_type = wine_main.wineDF.groupby(['variety'])
average_price = wine_type['price'].mean()
average_points = wine_type['points'].mean()
region_count = wine_type['points'].count()

wine_type_df = pd.DataFrame({
    'Average Price':average_price,
    'Average Points':average_points,
    'Wines in type':region_count
})

wine_type_dfn = wine_type_df.reset_index()

wine_type_dfn.describe()
```

```python
#Plot bubbles all wines
plt.scatter(wine_type_dfn['Average Points'],wine_type_dfn['Average Price'],
            color = 'red', edgecolor = 'white', alpha = .99)
plt.xlabel('Wine Score')
plt.ylabel('Wine Price')
plt.title('Wine Prices Vs Wine Score')
plt.tight_layout()
plt.savefig('winetypes-pirce-score.png')
plt.grid()

plt.show()
```

```python
#Drop wines above  $200 
wine_type_df_100 = wine_type_df[wine_type_df['Wines in type'] > 99].reset_index()
wine_type_df_300 = wine_type_df[wine_type_df['Wines in type'] > 299].reset_index()
wine_type_df_1000 = wine_type_df[wine_type_df['Wines in type'] > 999].reset_index()

#wine_type_df_100.head()
wine_type_df_1000.head()
```

```python
# Plot bubble 100 plus wines
plt.scatter(wine_type_df_100['Average Points'],wine_type_df_100['Average Price'],
            s = wine_type_df_100['Wines in type']/5,
            color = 'coral', edgecolor = 'black', alpha = .99)
plt.xlabel('Wine Score')
plt.ylabel('Wine Price')
plt.title('Wine Prices Vs Wine Score(100 plus wines)')
plt.tight_layout()
plt.grid()
plt.savefig('winetypes-pirce-score_100.png')
plt.show()
```

```python
#Plot Prices 100 plus wines
wine_type_df_100.sort_values('Average Price').plot.bar(x='variety',y='Average Price',figsize=(20,5),legend=False,rot=90)
plt.xlabel('Variety')
plt.ylabel('Average Wine Price')
plt.ylim(ymin=10)
plt.title('Average Wine Price per Variety (Over 100 Wines Reviewed)')
plt.tight_layout()
plt.savefig('Average_Wine_Prices_Variety_100.png')
plt.show()
```

```python
#Plot Scores 100 plus wines
wine_type_df_100.sort_values('Average Points').plot.bar(x='variety',y='Average Points',figsize=(20,5),legend=False,rot=90)
plt.xlabel('Variety')
plt.ylabel('Average Wine Score')
plt.ylim(ymin=85) 
plt.ylim(ymax=91) 
plt.title('Average Wine Score per Variety (Over 100 Wines Reviewed)')
plt.tight_layout()
plt.savefig('Average_Wine_Scores_Variety_100.png')
plt.show()
```

```python
# Plot bubble 300 plus wines
plt.scatter(wine_type_df_300['Average Points'],wine_type_df_300['Average Price'],
            s = wine_type_df_300['Wines in type']/5,
            color = 'gold', edgecolor = 'black', alpha = .99)
plt.xlabel('Wine Score')
plt.ylabel('Wine Price')
plt.title('Wine Prices Vs Wine Score(300 plus wines)')
plt.tight_layout()
plt.grid()
plt.savefig('winetypes-pirce-score_300.png')
plt.show()
```

```python
# Plot bubble 1000 plus wines 
plt.figure(figsize=(20,10))
plt.xlabel('Wine Score')
plt.ylabel('Wine Price')
plt.title('Wine Prices Vs Wine Score(1000 plus wines)')
plt.tight_layout()
plt.grid()
for index, row in wine_type_df_1000.iterrows():
    plt.scatter(row['Average Points'], row['Average Price'], s=row['Wines in type'], alpha=.7)
    plt.annotate(row['variety'], xy=(row['Average Points'],row['Average Price']))

    plt.savefig('winetypes-pirce-score_1000.png')
plt.show()
```
